### 1.4 Counting Vocabulary

 finding out the length of a text from start to finish

A __token__ is the technical name for a sequence of characters — such as  hairy, his, or :) — that we want to treat as a group.  
The vocabulary of a text is just the set of tokens that it uses, since in a set, all duplicates are collapsed together. In Python we can obtain the vocabulary items of text3 with the command: _set(text3)_.

```python
sorted(set(text3))
```

```python
len(set(text3))
```

A __word type__ is the form or spelling of the word independently of its specific occurrences in a text — that is, the word considered as a unique item of vocabulary. 

Now, let's calculate a measure of the lexical richness of the text. 

```python
len(set(text3)) / len(text3)
```

```python
text3.count("smote")
```

```python
100 * text4.count('a') / len(text4)
```

```python
def lexical_diversity(text):
    return len(set(text)) / len(text)
def percentage(count, total):
    return 100 * count / total
```

```python
lexical_diversity(text3)
```

```python
lexical_diversity(text5)
```

```python
percentage(4, 5)
```

```python
percentage(text4.count('a'), len(text4))
```

### 3.2   Fine-grained Selection of Words

```python
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

it would be better to find frequently occurring long words. This seems promising since it eliminates frequent short words (e.g., the) and infrequent long words (e.g. antiphilosophists).

```python
fdist5 = FreqDist(text5)
sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
```

### 4.4   Looping with Conditions

```python
sent1 = ['Call', 'me', 'Ishmael', '.']
for xyzzy in sent1:
    if xyzzy.endswith('l'):
        print(xyzzy)
```

```python
for token in sent1:
    if token.islower():
        print(token, 'is a lowercase word')
    elif token.istitle():
        print(token, 'is a titlecase word')
    else:
        print(token, 'is punctuation')
```

## 5   Automatic Natural Language Understanding

### 5.1   Word Sense Disambiguation  
In word sense disambiguation we want to work out which sense of a word was intended in a given context.

### 5.2   Pronoun Resolution  
A deeper kind of language understanding is to work out "who did what to whom" (to detect the subjects and objects of verbs)

### 5.3   Generating Language Output  
__question answering__: a machine should be able to answer a user's questions relating to collection of texts

__machine translation__: machine should be able to translate the text into another language, accurately conveying the meaning of the original text.

### 5.4   Machine Translation
Machine translation is difficult because a given word could have several possible translations (depending on its meaning), and because word order must be changed in keeping with the grammatical structure of the target language. 

### 5.5   Spoken Dialog Systems  
__Turing Test__: can a dialogue system, responding to a user's text input, perform so naturally that we cannot distinguish it from a human-generated response?  
![](./images/dialogue.png)

### 5.6   Textual Entailment  
Does the text provide enough evidence for you to accept the hypothesis?  
You can draw this conclusion easily, but it is very hard to come up with automated methods for making the right decision.
